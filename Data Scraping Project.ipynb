## Creating a Soup object

```python
if(res.status_code == 200):
    soup = bs(res.text,'html.parser')
else:
    print('Some Error has occured please check it out')
```

```python
alltags = soup.tagStack
alltags
```

```python
alllielements = soup.find_all('li')
alllielements
```

## Scrape the title and prices of book

```python
itemscontainer = soup.find('ol',class_='row')
```

```python
childelems = itemscontainer.find_all(recursive=False)
```

```python
allbooktitles = [titles.find('h3').text for titles in childelems]
```

```python
allbooktitles
```

```python
allbookprices = [elem.find('p',class_='price_color').text for elem in childelems]
```

```python
allbookprices
```

## All book details

```python
allbooksdata = []
for elems in childelems:
    if (elems.find('p',class_='instock').text.strip() == "In stock"):
        isavailable = 'available'
    else:
        isavailable = 'not available'
        
    productdetails = {
        "Title":"{}".format(elems.find('h3').text),
        "Price":"{}".format(elems.find('p',class_='price_color').text),
        "Availablitiy":"{}".format(isavailable),
        "Ratings":"{}".format(elems.find('p','star-rating').get('class')[1]), # 1st index of 2 classes
#         "Product Description":"{}".format(product_desc)
    }

    allbooksdata.append(productdetails)
    
allbooksdata
```

## Convert the book Data into .csv file

```python
import csv
```

```python
csv_file_path = 'books_data.csv'

with open(csv_file_path,'w',newline='',encoding='utf-8') as csv_file:
    csv_writer = csv.DictWriter(csv_file,fieldnames=allbooksdata[0].keys())
    csv_writer.writeheader()
    csv_writer.writerows(allbooksdata)
    
print(f'CSV file {csv_file_path} has been created successfully')
```
